```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = LinearRegression()
model
```

```python
apartamente = np.array([
   (30, 40_000), (45, 65_000), (55, 130_000),  (60, 145_000), (75, 180_000), (75, 185_000), (80, 200_000), (100, 300_000), # (80, 40_000)
])
suprafete = apartamente[ : , 0 ]
preturi = apartamente[ : , 1 ]
```

```python
suprafete
```

```python
suprafete_2d = suprafete.reshape(-1, 1)
suprafete_2d
```

```python
model.fit(suprafete_2d, preturi)
```

```python
plt.scatter(suprafete, preturi);
# plt.plot(suprafete, preturi);
plt.plot(suprafete, model.coef_ * suprafete + model.intercept_, c="black");
plt.xlabel("Suprafete");
plt.ylabel("Preturi");
```

```python
model.predict([ [40] ])
```
