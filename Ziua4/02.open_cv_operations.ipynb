```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
cat_img = cv2.imread('cat.jpg')
```

```python
plt.imshow(cat_img)
```

```python
plt.imshow(cat_img, cmap="gray")
```

```python
gray_cat_img = cv2.imread('cat.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(gray_cat_img, cmap="gray")
```

```python
gray_cat_img.shape
```

```python
resized_image = cv2.resize(gray_cat_img, (gray_cat_img.shape[1] // 3, gray_cat_img.shape[0] // 3) )
plt.imshow(resized_image, cmap="gray")
```

```python
top_sobel = np.array([1, 2, 1, 0, 0, 0, -1, -2, -1]).reshape(3, 3)
top_sobel
```

```python
top_sobel_image = cv2.filter2D(gray_cat_img,-1,top_sobel)
plt.imshow(top_sobel_image, cmap="gray")
```

```python
left_sobel = np.array([1, 0, -1, 2, 0, -2, 1, 0, -1]).reshape(3, 3)
left_sobel_image = cv2.filter2D(gray_cat_img,-1,left_sobel)
plt.imshow(left_sobel_image, cmap="gray")
```
