# 1. Map 

```python
o_lista = [23, 34, 42, 321, 321, 312, 553, 2123, 335]


def scadere(x):
    return x / 2


list(map(scadere, o_lista))
```

```python
o_lista = [23, 34, 42, 321, 321, 312, 553, 2123, 335]


[ i / 2 for i in o_lista ]
```

# 2. Filter

```python
o_lista = [23, 34, 42, 321, 321, 312, 553, 2123, 335]

def mai_mic_ca_100(x):
    return x  < 100


list(filter(mai_mic_ca_100, o_lista))
```

```python
o_lista = [23, 34, 42, 321, 321, 312, 553, 2123, 335]

list(filter(lambda x : x < 100 , o_lista))
```

```python
o_lista = [23, 34, 42, 321, 321, 312, 553, 2123, 335]

[ i for i in o_lista if i < 100 ]
```

# 3. Reduce

```python
from functools import reduce

o_lista = [23, 34, 42, 321, 321, 312, 553, 2123, 335]


def cel_mai_mare(x, y):
    return x if x < y else y


reduce(cel_mai_mare, o_lista)
```

```python
cel_mai_mare(10, 20)
```

```python
o_lista = [23, 34, 42, 321, 321, 312, 553, 2123, 335]
reduce(lambda x, y: x if x < y else y, o_lista)
```
