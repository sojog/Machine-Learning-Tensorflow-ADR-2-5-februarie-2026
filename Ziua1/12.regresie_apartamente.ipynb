```python
from scipy import stats
```

```python
lingregress_result = stats.linregress(suprafete, preturi)
lingregress_result
```

```python
# a                          b
lingregress_result.slope, lingregress_result.intercept
```

```python
def f_perfect (x):
    return lingregress_result.slope * x + lingregress_result.intercept
```

```python
suprafete
```

```python
np.array(list(map(f_perfect, suprafete)))
```

```python
lingregress_result.slope * suprafete + lingregress_result.intercept
```

```python
plt.scatter(suprafete, preturi);
plt.plot(suprafete, preturi);
plt.plot(suprafete, lingregress_result.slope * suprafete + lingregress_result.intercept, c="black");
```

```python
plt.scatter(suprafete, preturi);
# plt.plot(suprafete, preturi);
plt.plot(suprafete, lingregress_result.slope * suprafete + lingregress_result.intercept, c="black");
plt.xlabel("Suprafete");
plt.ylabel("Preturi");
```
