```python
x= np.array([8, 6, 5, 4, 5, 6, 7, 9, 5, 5])
y= np.array([5, 1, 4, 4, 9, 1, 8, 5, 1, 6])

zipped_array = np.array(list(zip(x, y)))
zipped_array
```

```python
sorted_array = np.array(sorted(zipped_array, key=lambda x:x[0]))
sorted_array
```

```python
plt.plot(x, y);
```

```python
plt.plot(sorted_array[:, 0], sorted_array[:, 1]);
```
