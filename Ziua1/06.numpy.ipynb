# Tuple

```python
t = ()
print(t, type(t))
```

```python
t = (2, 3, 4)
print(t, type(t))
```

```python
t = (2)
print(t, type(t))
```

```python
t = ((((2))))
print(t, type(t))
```

```python
t = ((((2,))))
print(t, type(t))
```

```python
t = 2,
print(t, type(t))
```

```python
t = 2, 3
print(t, type(t))
```

```python
# unpacking
x, y = t
print("x=", x)
print("y=", y)
```

```python
x, y, z, *a = 2, 3, 4
print("x=", x)
print("y=", y)
print("z=", z)
print("a=", a)
```

```python
x, y, z, *a = 2, 3, 4, 7, 8, 8
print("x=", x)
print("y=", y)
print("z=", z)
print("a=", a)
```

```python
x = 20
y = "helllo"

x, y = y, x
print("x=", x)
print("y=", y)
```
