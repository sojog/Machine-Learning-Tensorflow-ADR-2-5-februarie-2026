```python
from sklearn.model_selection import train_test_split
```

```python


x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_label = 'sepal length (cm)'
y_label = 'sepal width (cm)'

plt.scatter(x=x_train[ : , 0], y=x_train[:,1])
plt.scatter(x=x_test[ : , 0], y=x_test[:,1])

plt.xlabel(x_label)
plt.ylabel(y_label)
plt.legend(iris_df["target_name"])
```

```python
x_label = 'petal length (cm)'
y_label = 'petal width (cm)'

plt.scatter(x=x_train[ : , 2], y=x_train[:,3])
plt.scatter(x=x_test[ : , 2], y=x_test[:,3])

plt.xlabel(x_label)
plt.ylabel(y_label)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model = GridSearchCV(KNeighborsClassifier(), {'n_neighbors':range(2, 11)})
model
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.best_estimator_.n_neighbors
```
