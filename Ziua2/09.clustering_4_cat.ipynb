```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_4_cat = pd.read_csv("categories_datasets/4_categories.csv")
df_4_cat
```

```python
df_4_cat.category.unique()
```

```python
plt.scatter(df_4_cat["coord_x"], df_4_cat["coord_y"], c=df_4_cat.category.map({'a':'red', 'b':'green', 'c':'black', 'd':'pink'}))
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans_2_cat = KMeans(n_clusters=4)
kmeans_2_cat
```

```python
kmeans_2_cat.fit(df_4_cat[["coord_x", "coord_y"]])
```

```python
predicted_categories = kmeans_2_cat.predict(df_4_cat[["coord_x", "coord_y"]])
predicted_categories
```

```python
plt.scatter(df_4_cat["coord_x"], df_4_cat["coord_y"], c=df_4_cat.category.map({'a':'red', 'b':'green', 'c':'black', 'd':'pink'}))
```

```python
plt.scatter(df_4_cat["coord_x"], df_4_cat["coord_y"], c=predicted_categories)
```
