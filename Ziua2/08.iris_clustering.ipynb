```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
iris
```

```python
iris.keys()
```

```python
iris.DESCR
```

```python
print(iris["DESCR"])
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df
```

```python
iris_df['target'] = iris.target
iris_df
```

```python
iris.target_names
```

```python
iris_df['target_name'] = iris_df["target"].map({0:'setosa', 1:'versicolor', 2:'virginica'})
iris_df
```

```python
plt.scatter(x=iris_df['sepal length (cm)'], y=iris_df['sepal width (cm)'])
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
```

```python
x_label = 'petal length (cm)'
y_label = 'petal width (cm)'

plt.scatter(x=iris_df[x_label], y=iris_df[y_label])
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.legend(iris_df["target_name"])
```

# Clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans_model = KMeans(n_clusters=3)
kmeans_model
```

```python
x = iris.data
```

```python
kmeans_model.fit(x)
```

```python
y_prezis = kmeans_model.predict(x)
y_prezis
```

```python
x_label = 'petal length (cm)'
y_label = 'petal width (cm)'

plt.scatter(x=iris_df[x_label], y=iris_df[y_label], c=y_prezis)
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.legend(iris_df["target_name"])
```

```python
x_label = 'petal length (cm)'
y_label = 'petal width (cm)'

plt.scatter(x=iris_df[x_label], y=iris_df[y_label], c=iris.target)
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.legend(iris_df["target_name"])
```
