```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier()
dt_model
```

```python
df_2_cat
```

```python
x = df_2_cat[["coord_x", "coord_y"]]
y = df_2_cat["category"]
```

```python
dt_model.fit(x, y)
```

# pip install graphviz

```python
from sklearn.tree import export_graphviz
import graphviz
```

```python
exported_data = export_graphviz(dt_model, feature_names=["coord_x", "coord_y"],class_names=['a', 'b'], filled=True, rounded=True)
exported_data
```

```python
tree_graph = graphviz.Source(exported_data)
tree_graph.render("2_categories", format="png")
tree_graph.view()
```
