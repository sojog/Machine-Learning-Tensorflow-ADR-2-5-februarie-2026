```python
x = df [ ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']]
x
```

```python
x = df.drop('MEDV', axis=1)
x
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
model = LinearRegression()
model
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model.fit(x_train, y_train)
```

```python
y_prezis = model.predict(x_test)
y_prezis
```

```python
y_test
```

# Comparatie y_prezis (x_test) vs y_test

```python
comp_df = pd.DataFrame({ "prezis": y_prezis.flatten(),
"real": y_test["MEDV"] })
comp_df
```

```python
plt.scatter( y_test["MEDV"] ,  y_prezis.flatten())
plt.xlabel("Y real")
plt.ylabel("Y prezis")
```

```python
comp_df["score"] = (comp_df["prezis"]- comp_df["real"]) ** 2
comp_df
```

```python
comp_df["score"].sum()
```

```python
np.sqrt(comp_df["score"].sum())
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_model = KNeighborsRegressor()
knn_model
```

```python
knn_model.fit(x_train, y_train)
```

```python
knn_model.score(x_test, y_test)
```

```python
knn_scores = {}

for k in range(2, 11):
    knn_model = KNeighborsRegressor(n_neighbors=k)
    knn_model.fit(x_train, y_train)
    score = knn_model.score(x_test, y_test)
    knn_scores[f"{k}"] = score

```

```python
knn_scores
```

```python
np.array(list(knn_scores.values())).max()
```

```python
np.array(list(knn_scores.values())).argmax()
```
