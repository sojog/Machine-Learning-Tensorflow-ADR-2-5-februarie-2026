```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Source : https://www.kaggle.com/code/prasadperera/the-boston-housing-dataset

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df = pd.read_csv('Boston_data.csv', header=None, delimiter=r"\s+", names=column_names)
df
```

```python
df.keys()
```

```python
df.describe()
```

```python
y = df [['MEDV']]
y
```

```python
type(y)
```

```python
x = df [ ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']]
x
```

```python
x = df.drop('MEDV', axis=1)
x
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
        # ('nume_pas', estimator)
linear_steps = [('scaler', StandardScaler()), ('model', LinearRegression())]
```

```python
linear_pipeline = Pipeline(linear_steps)
linear_pipeline
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
linear_pipeline.fit(x_train, y_train)
```

```python
y_prezis = linear_pipeline.predict(x_test)
y_prezis
```

```python
y_test
```

# Comparatie y_prezis (x_test) vs y_test

```python
comp_df = pd.DataFrame({ "prezis": y_prezis.flatten(),
"real": y_test["MEDV"] })
comp_df
```

```python
plt.scatter( y_test["MEDV"] ,  y_prezis.flatten())
plt.xlabel("Y real")
plt.ylabel("Y prezis")
```

```python
comp_df["score"] = (comp_df["prezis"]- comp_df["real"]) ** 2
comp_df
```

```python
comp_df["score"].sum()
```

```python
np.sqrt(comp_df["score"].sum())
```

```python
linear_pipeline.score(x_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
        # ('nume_pas', estimator)
knn_steps = [('scaler', StandardScaler()), ('model', KNeighborsRegressor())]
```

```python
knn_pipeline = Pipeline(knn_steps)
knn_pipeline
```

```python
knn_pipeline.fit(x_train, y_train)
```

```python
knn_pipeline.score(x_test, y_test)
```

```python
x_test.iloc[0]
```

```python
knn_pipeline.predict([x_test.iloc[0]])
```

```python
knn_pipeline["model"]
```

```python
knn_pipeline["model"].n_neighbors 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
GridSearchCV(KNeighborsRegressor(), {
    'n_neighbors' : range(2, 11)
})
```

```python
parameter_grid = {
    'model__n_neighbors' : range(2, 11)
}

knn_grid_model = GridSearchCV(knn_pipeline, parameter_grid)
knn_grid_model
```

```python
knn_grid_model.fit(x_train, y_train)
```

```python
knn_grid_model.best_estimator_
```

```python
knn_grid_model.best_estimator_["model"]
```

```python
knn_grid_model.best_estimator_["model"].n_neighbors
```

```python
knn_grid_model.score(x_test, y_test)
```
