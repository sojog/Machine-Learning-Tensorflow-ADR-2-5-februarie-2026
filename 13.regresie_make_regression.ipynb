```python
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
regression = make_regression(n_samples=2000, n_features=1, noise=20)
type(regression), len(regression)
```

```python
x, y = regression
x.shape, y.shape
```

```python
from scipy import stats

results = stats.linregress(x.flatten(), y)
results
```

```python
plt.scatter(x, y)
plt.plot(x, x * results.slope + results.intercept);
```
