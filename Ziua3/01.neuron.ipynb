# Rezulatatul neuronul as vrea sa fie in intervalul 0..1    (un numar mai mic)

```python
def functie_activare(valoarea):
    return 1 if valoarea > 0 else 0

def AND_neuron (x1, x2):
    w1 = 1 # trebuie gasite
    w2 = 1 
    b = -1
    rezultat =  functie_activare(x1 * w1 + x2 * w2 + b)
    print(x1, "AND", x2, "=", rezultat)
    
AND_neuron(1, 1)
AND_neuron(1, 0)
AND_neuron(0, 1)
AND_neuron(0, 0)
```

# Similar este cu neuronul OR

```python
def functie_activare(valoarea):
    return 1 if valoarea >= 0 else 0

def OR_neuron (x1, x2):
    w1 = 1 # trebuie gasite
    w2 = 1 
    b = -0.5
    rezultat =  functie_activare(x1 * w1 + x2 * w2 + b)
    print(x1, "AND", x2, "=", rezultat)
    
OR_neuron(1, 1)
OR_neuron(1, 0)
OR_neuron(0, 1)
OR_neuron(0, 0)
```
