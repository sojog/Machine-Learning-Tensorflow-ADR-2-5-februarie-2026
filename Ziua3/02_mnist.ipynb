```python
model = tf.keras.Sequential(
    [tf.keras.layers.Input(shape=(784,))]
)
```

```python
model = tf.keras.Sequential(

     [
         ## INPUT LAYER -  va fi intotdeauna la fel, in functie de tipul datelor
         tf.keras.layers.Flatten(input_shape=(28, 28)),


         tf.keras.layers.Dense(128, activation="relu"),

         #tf.keras.layers.Dense(64, activation="relu"),


          ## OUTPUT LAYER - va fi intotdeauna la fel in functie de ceea ce genereaza
          tf.keras.layers.Dense(10, activation="softmax")


     ],
)
```

```python
# V1 - versiunea declarativa
model.compile(loss='sparse_categorical_crossentropy',
             optimizer='adam', metrics=['accuracy'])
```

```python
# V2 - versiunea explicita
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),  metrics=['accuracy'] )
```

```python
model_history = model.fit(x_train_scaled, y_train, epochs=10)
```

```python
x_test_scaled = x_test / 255.0

test_loss, test_acc = model.evaluate(x_test_scaled, y_test)
```
