```python
df = pd.read_csv("Titanic-Dataset.csv", index_col=0)
df
```

```python
df = df.drop("Name", axis=1)
df.head(2)
```

```python
df["Sex"].unique()
```

```python
df["Gen"] = df["Sex"].map({"male":0, "female":1})
df.head()
```

```python
df.describe()
```

```python
df.drop("Sex", axis=1, inplace=True)
df
```

```python
df["Embarked"].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder, MultiLabelBinarizer, LabelBinarizer, LabelEncoder
```

```python
df["Embarked_mapped"] = LabelEncoder().fit_transform(df["Embarked"])
df
```

```python
df["Embarked_mapped"].unique()
```
